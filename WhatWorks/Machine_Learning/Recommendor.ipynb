```python
import numpy as np
import pandas as pd
import random

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

# Custom Functions
def combined_features(row):
    return row['Tags']+" "+row['User']

# Now we create our custom data
tag_data = [];
username_data = [];

# Create 2000 tag data's
for x in range(0,2000):
  # Pre-made tags
  temp = ["highfat" + " " + "greasy"];
  for y in range(0,random.randint(1, 5)):
    temp[0] = temp[0] + " " + food_tags[random.randint(0, 62)] + " ";
    temp[0] = temp[0] + "triggering";
  tag_data.append(temp);

# Create 2000 users
for x in range(0, 2000):
  temp =userArray[random.randint(0, 6)];
  username_data.append(temp);

# print(tag_data);
print(username_data);

df = pd.DataFrame(tag_data, columns =['Tags']); 

df.insert(1, "User", username_data);
df["combined_features"] = df.apply(combined_features, axis =1)
print(df);
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
# print("Count Matrix:", count_matrix.toarray())

cosine_sim = cosine_similarity(count_matrix)

# print(cosine_sim);

similar_reviews = list(enumerate(cosine_sim[300]));

sorted_similar_review = sorted(similar_reviews, key=lambda x:x[1], reverse=True)

print(sorted_similar_review);
print("THE TARGET TAG: ", tag_data[300]);
print("THE TARGET USER: ", username_data[300]);
print()
print("THE PROPOSED TAG: ", tag_data[1692]);
print("THE PROPOSED USER: ", username_data[1692]);
```
